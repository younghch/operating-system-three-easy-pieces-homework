```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
columns = ["numberOfPages", "trials", "seconds"]
df2000 = pd.read_csv("result-2000.csv", names=columns)
df4000 = pd.read_csv("result-4000.csv", names=columns)
df8000 = pd.read_csv("result-8000.csv", names=columns)
```

```python
def plot_line(df, n, title, color):
    fig = plt.figure(figsize=(5, 10))
    x = list(df.numberOfPages)[:n]
    y = list(df.seconds)[:n]
    plt.plot(x, y, marker='o', color=color)
    plt.xscale('symlog', base=2)
    plt.xlabel('number of pages')
    plt.ylabel('s')
    plt.title(title)
    plt.show()
```

```python
plot_line(df2000, 10, '2000 trials', 'orange')
plot_line(df4000, 10, '4000 trials', 'red')
plot_line(df8000, 10, '8000 trials', 'black')
```
